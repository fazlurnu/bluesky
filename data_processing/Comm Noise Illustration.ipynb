```python
import numpy as np

size = 400
stdev = 3

last_update = np.zeros(size) + abs(np.random.normal(0, stdev, size = 400))
# print(last_update)
simt = 0

printed_once = False

last_update_all = []

for i in range(10001):
    simt += 1
    
    up = np.where(last_update < simt)
        
    last_update = last_update - abs(np.random.normal(0, stdev, size = 400))
    last_update[up] = np.full(len(last_update[up]), simt) - abs(np.random.normal(0, stdev, size = len(last_update[up])))
    
    last_update_all.extend(last_update[up]-simt)
    
plt.hist(last_update_all, bins=100);
#     print(up)
    
# print(last_update)
```
